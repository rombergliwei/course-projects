```python
#使用opencv
#将同一张图片复制拼贴
#图片像素点个数不同不能拼贴

import cv2
from pylab import *

img1 = cv2.imread('1.jpg',cv2.IMREAD_COLOR)
img2 = cv2.imread('1.jpg',cv2.IMREAD_GRAYSCALE)
img3 = cv2.imread('1.jpg',cv2.IMREAD_UNCHANGED)
img4 = cv2.imread('1.jpg')

htitch= np.hstack((img1, img3,img4))
vtitch = np.vstack((img1, img3))
cv2.imshow("test1",htitch)
cv2.imshow("test2",vtitch)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
#from pylab import *

img1 = cv2.imread('1.jpg')
img2 = cv2.imread('2.jpg')
img3 = cv2.imread('3.jpg')
img4 = cv2.imread('4.jpg')
img5 = cv2.imread('5.jpg')

# 将opencv中的BGR、GRAY格式转换为RGB，使matplotlib中能正常显示opencv的图像

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2RGB)

fig = plt.figure()
subplot(331)
imshow(img1)
title('img1')
axis('off')
subplot(332)
imshow(img2)
title('img2')
axis('off')
subplot(333)
imshow(img3)
title('img3')
axis('off')
subplot(334)
imshow(img4)
title('img4')
axis('off')
subplot(335)
imshow(img5)
title('img5')
axis('off')
show()
```
