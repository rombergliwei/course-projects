# 1. Write a function to automatically load all the images (use 5 images with different sizes/dimensions) in a folder

```python
import os
from PIL import Image
import numpy as np
import cv2
array_of_img = [] # this if for store all of the image data
# this function is for read image,the input is directory name
def loadImage(directory_name):
    # this loop is for read each image in this foder,directory_name is the foder name with images.
    for filename in os.listdir(directory_name):
        #print(filename) #just for test
        #imgs is used to store the image data 
        imgs = cv2.imread(directory_name + "/" + filename)
        array_of_img.append(imgs)
        #print(img)
        print(array_of_img)
path="C:/Users/weilu/dikong/picture1"
loadImage(path)
```

# 3. Data augmentation:
* Rotate your images images by 90 degrees, display the rotated images
* Flip your images horizontally, display the flipped images
* Flip your images vertically, display the flipped images

```python
# you codes for data augmentation go here
import os
import numpy as np
import cv2

path="C:/Users/weilu/dikong/picture1"
for filename in os.listdir(path):
        img = cv2.imread(path + "/" + filename)
        img90=np.rot90(img)
        cv2.imshow("rotate",img90)
        cv2.waitKey(0)


```

```python
import os
import numpy as np
import cv2

path="C:/Users/weilu/dikong/picture1"
for filename in os.listdir(path):
        img = cv2.imread(path + "/" + filename)
        xImg = cv2.flip(img,1,dst=None)
        cv2.imshow("horizontal",xImg)
        cv2.waitKey(0)
```

# 5. Resize and display
## use the python package of `cv2` to resize all the images into the same size/shape of (256,256,3), and display the resized images

```python
# install cv2 first if you have not done that before
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt 
from pylab import *

path="C:/Users/weilu/dikong/picture1"
for filename in os.listdir(path):
        img = cv2.imread(path + "/" + filename)
        res=cv2.resize(img,(256,256),interpolation=cv2.INTER_CUBIC)
        cv2.imshow("resize",res)
        cv2.waitKey(0)

```
